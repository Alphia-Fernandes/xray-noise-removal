```python
import matplotlib.pyplot as plt
import cv2
import os

noisy_folder = "noisy_images"
denoised_folder = "denoised_output"
os.makedirs(denoised_folder, exist_ok=True)

image_name = "1.png"  # Change this to test other noisy images
noisy_path = os.path.join(noisy_folder, image_name)
denoised_path = os.path.join(denoised_folder, image_name)

if os.path.exists(noisy_path):
    noisy_image = cv2.imread(noisy_path, cv2.IMREAD_GRAYSCALE)

    # Apply median filter for denoising salt-and-pepper noise
    denoised_image = cv2.medianBlur(noisy_image, ksize=3)

    cv2.imwrite(denoised_path, denoised_image)

    # Show just noisy and denoised images
    plt.figure(figsize=(10, 5))
    titles = ['Noisy (Salt & Pepper)', 'Denoised (Median Filter)']
    images = [noisy_image, denoised_image]

    for i in range(2):
        plt.subplot(1, 2, i+1)
        plt.imshow(images[i], cmap='gray')
        plt.title(titles[i])
        plt.axis('off')

    plt.tight_layout()
    plt.show()

else:
    print(f" Image {image_name} not found in {noisy_folder}/")
```
